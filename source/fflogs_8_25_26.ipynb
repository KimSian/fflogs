```python
plt_count = 1
for f in floor:
    data = pd.read_csv(r'C:\Users\YH\PycharmProjects\untitled\fflogs\preprocessing\user_info\user_info_'+f+'.txt', engine='python', encoding='utf-8')
    
    plt.subplot(4,1,plt_count)

    ax = sns.countplot(x = data['직업'], data=data, order=data['직업'].value_counts().index)
    plt.title('Job Count - First Floor')
    plt.ylabel('Population')
    plt.xlabel('Job')
    for p in ax.patches:
        ax.annotate('{:.0f}'.format(p.get_height()), (p.get_x()+0.3, p.get_height()+10))
    
    plt_count += 1
    
plt.tight_layout()
```

```python
# 3-1) 인원 증감에 대한 꺾은선 그래프

job_count_floor = pd.DataFrame(columns=['BlackMage', 'Monk', 'Dragoon', 'Bard', 'DarkKnight', 'Paladin',
       'WhiteMage', 'Scholar', 'Dancer', 'Warrior', 'RedMage',
       'Gunbreaker', 'Machinist', 'Samurai', 'Astrologian', 'Summoner',
       'Ninja'])

names = ['BlackMage', 'Monk', 'Dragoon', 'Bard', 'DarkKnight', 'Paladin',
       'WhiteMage', 'Scholar', 'Dancer', 'Warrior', 'RedMage',
       'Gunbreaker', 'Machinist', 'Samurai', 'Astrologian', 'Summoner',
       'Ninja']
job_count_floor.reindex([1,2,3,4])
floor_count = 1
for f in floor:
    data = pd.read_csv(r'C:\Users\YH\PycharmProjects\untitled\fflogs\preprocessing\user_info\user_info_'+f+'.txt', engine='python', encoding='utf-8')
    job_count = data['직업'].value_counts()
    for name in names:
        job_count_floor.loc[floor_count,name]= job_count[name]
    floor_count += 1
job_count_floor
```

```python
plt.plot(figsize=(10,9))
ax = plt.subplot(1,1,1)
ax.set_xticks([1,2,3,4])
plt.plot(job_count_floor)
plt.ylabel('인원')
plt.xlabel('층')
plt.legend(names)
```

```python
#층 별 현황
job_count_floor_T = job_count_floor.T

plt.plot(figsize=(10,9))
plt.plot(job_count_floor_T)
plt.ylabel('Population')
plt.xlabel('Job')
plt.legend(['First', 'Second', 'Third', 'Fourth'])
```
